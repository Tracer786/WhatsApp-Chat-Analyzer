```python
#Preprocessing

#We have all the Chats in a text file.
#Will extract all the Chats in form of text and convert it into a Pandas Dataframe.
```

```python
import re #regular expression
import pandas as pd #importing the pandas library
```

```python
f = open('Niggas.txt','r',encoding='utf-8')
```

```python
data = f.read() #read all the data from the .txt file in form of a string
```

```python
print(data)
```

```python
print(type(data)) #to check the type of data (whether it is string type or not)
```

```python
#create a new dataframe containing 2 separate columns 
#column 1 for the date
#column 2 for the name and message

pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
        #first \d{1,2} represents that the regular expression will check the expression for digits and the length of the digits can be 1 or 2
        #then there is a '/'
        #second \d{1,2} represents the same as the first d
        #then there is a '/'
        #second \d{2,4} represents the same as the first d but for the year
        #then we have a comma ','
        #after that a space '\s'
        #then similarly for the time
        
        
# pattern = r'\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s[APMapm]{2}\s-\s'

        
```

```python
messages = re.split(pattern,data)[1:]  #the 1: means the messages from the 1 onwards. Not print the first message i.e. the[0] message
messages



# result = re.split(pattern, data)

# if len(result) > 1:
#     messages = result[1]
#     # Your further processing with messages
# else:
#     # Handle the case when there are no elements or only one element in the result
#     messages = None  # Or set it to an appropriate default value or raise an error
    
# result
```

```python
dates = re.findall(pattern, data)
dates
#extracting the dates from the message list
```

```python
#creating a pandas dataframe
#the name of the first column is user_message and the name of the second column is message_date 
#first column will contain the list of messages and the second column will contain the list of dates
df = pd.DataFrame({'user_message': messages, 'message_date': dates})
#converting message_date datatype
df['message_date'] = pd.to_datetime(df['message_date'],format='%d/%m/%Y, %H:%M - ')
df.rename(columns={'message_date': 'date'},inplace=True)
df.head()


#user ka name and message ek single colum me aa rhe hai
```

```python
#to check the total number of messages
df.shape
```

```python
#separate users and messages from the first column i.e. the user_message column
users = []
messages = []
#creating lists for users and messages
for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s', message)
    if entry[1:]: #user name
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(entry[0])
    #means agar kisi message me ek name hai and uske baad ek colon (:) hai and then uske baad ek string message hai to hm ek nya list create krenge users name ka and usme username ko append kr denge
    #aur second part of the string jo colon ke baad hai use hm messages wali list me append  kr denge
    
    
#creating the new columns
df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'],inplace = True)
    
df.head()
```

```python
period = []
for hour in df[['day_name', 'hour']]['hour']:
    if hour == 23:
        period.append(str(hour) + "-" + str('00'))
    elif hour == 0:
        period.append(str('00') + "-" + str(hour+1))
    else:
        period.append(str(hour) + "-" + str(hour+1))
```

```python
df['period'] = period
```
